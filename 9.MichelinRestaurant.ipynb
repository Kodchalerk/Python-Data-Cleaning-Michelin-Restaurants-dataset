```python
# Change columns' names to lowercase
df.columns = [i.lower().replace(' ','_') for i in df.columns]
```

```python
df.columns
```

```python
# Drop unnecessary columns
df.drop(columns = ['address', 'phonenumber', 'url', 'websiteurl'],inplace = True)
```

```python
df
```

```python
fac = df['facilitiesandservices'].str.get_dummies(',')

def impute(x):
    if x == 0:
        return 'No'
    else:
        return 'Yes'
    
for k in fac.columns:
    fac[k] = fac[k].apply(impute)
    

fac.columns = [i.lower().replace(' ','_') for i in fac.columns]
```

```python
df = pd.concat([df,fac], axis = 1)
```

```python
df.drop(columns = 'facilitiesandservices' , inplace = True)
```

```python
# Check for null values
print('Null Values')
print('-------------------')
print(df.isnull().sum())
print('-------------------')
print('Null Values Percentage')
print('-------------------')
print(df.isnull().mean()*100)
```

```python
# Check for duplicated values
df.duplicated().sum()
```

```python
# Drop null values
df.dropna(inplace = True)
```

```python
# Deal with Price
df['price'].unique()
```

```python
df['price'] = df['price'].astype('str')
```

```python
def price(x):
    if len(x) == 4:
        return 'Very Expensive'
    elif len(x) == 3:
        return 'Expensive'
    elif len(x) == 2:
        return 'Moderately expensive'
    else:
        return 'inexpensive'

df['price'] = df['price'].apply(price)
```

```python
# Now the data is clean
location = df['location'].str.split(',',expand = True)
location['location1'] = location[1] + ',' + location[0]
location['location1'] = np.where(location['location1'].isnull() , location[0] , location['location1'])
```

```python
location = location['location1']
```

```python
df = pd.concat([df,location], axis = 1)
```

```python
df.drop(columns = 'location', inplace = True)
```
